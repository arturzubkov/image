```python
import math
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cv2.getGaborKernel(ksize, sigma, theta, lambda, gamma, psi, ktype)
# ksize = size of gabor filter (n,n)
# sigma = standart  deviation of the gaussian function
# theta = orientation of the sinusoidal factor
# gamma = spatial aspect ratio 
# psi = phase offset
# ktype = type and range of values that each pixel in the gabor kernel can hold
```

```python
No = 4
Nw = 4
N = No*Nw
n = 0
plt.figure(figsize = (20,20))
for j in range(Nw): # wavelength
    for i in range(No): # orientation
        kernel = cv2.getGaborKernel((21,21), 5, np.pi/4*i, 4*(j+1),1,0,cv2.CV_32F)
        kernel /= math.sqrt((kernel*kernel).sum())
        n = n+1
        plt.subplot(Nw,No,n)
        plt.imshow(kernel, cmap="rainbow")
plt.show()
```

```python
# Using the filter with an image

my_image = cv2.imread("goedel.jpg", 0).astype(np.float32)/255
cv2.imshow("image", my_image)
Nw = 4
N = No*Nw
n=0

plt.figure(figsize=(25,25))
for j in range(Nw): # wavelength
    for i in range(No): # orientation
        kernel = cv2.getGaborKernel((26,26), 3, np.pi/3*i, 4*(j+1), 1,0, cv2.CV_32F)
        kernel /= math.sqrt((kernel*kernel).sum())
        filtered = cv2.filter2D(my_image, -1, kernel)
        n = n+1
        plt.subplot(Nw,No,n)
        plt.imshow(filtered, cmap="rainbow")
        plt.axis("off")
plt.show()
```

```python
me = cv2.imread("me.jpg", 0).astype(np.float32)
cv2.imshow("me", me)
Nw = 4
N = No*Nw
n=0
plt.figure(figsize=(20,20))
for j in range(Nw):
    for i in range(No):
        kernel = cv2.getGaborKernel((21,21), 4, np.pi/5*i, 4*(j+1), 1,0,cv2.CV_32F)
        kernel /= math.sqrt((kernel*kernel).sum())
        filtered = cv2.filter2D(me, -1, kernel)
        n = n + 1
        plt.subplot(Nw,No,n)
        plt.imshow(filtered, cmap="rainbow")
        plt.axis("off")
plt.show()

```
